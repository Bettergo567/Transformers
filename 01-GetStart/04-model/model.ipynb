```python
sen = '我有一个大梦想'
tokenizer = AutoTokenizer.from_pretrained("hfl/rbt3")
inputs = tokenizer(sen,return_tensors='pt')
inputs
```

### 不带model head的模型调用

```python
model = AutoModel.from_pretrained("hfl/rbt3", output_attentions=True)
```

```python
output = model(**inputs)
output
```

```python
output.last_hidden_state.size()
```
