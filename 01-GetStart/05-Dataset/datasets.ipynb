```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")
```

```python
def process_fun(example,tokenizer=tokenizer):
    model_inputs = tokenizer(example["content"],max_length=512,truncation=True)
    labels = tokenizer(example["title"],max_length=32,truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
datasets = load_dataset("madao33/new-title-chinese")
```

```python
processed_datasets = datasets.map(process_fun)
```

```python
processed_datasets
```

```python
processed_datasets = datasets.map(process_fun, num_proc=4) #num——proc使用四个进程加快
processed_datasets
```

```python
processed_datasets = datasets.map(process_fun, batched=True,remove_columns=datasets['train'].column_names) 
processed_datasets
```
